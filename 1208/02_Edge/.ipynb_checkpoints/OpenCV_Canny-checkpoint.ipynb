# OpenCV_Canny

```python
import numpy as np
import cv2 as cv
```

```python
#載入Lena範例影像，並將其轉換為灰階影像
img_lena = cv.imread("./img/lena.png")
gray_lena = cv.cvtColor(img_lena, cv.COLOR_BGR2GRAY)

#載入鑰匙範例影像，並將其轉換為灰階影像
img_key = cv.imread("./img/key.png")
gray_key = cv.cvtColor(img_key, cv.COLOR_BGR2GRAY)

#顯示影像處理結果並與原圖比較
cv.imshow('Ori_key', img_key)  
cv.imshow('key_gray',gray_key)
cv.imshow('Ori_Lena',img_lena)  
cv.imshow('lena_gray',gray_lena)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Canny 下界門檻值為 50;上界門檻值為 150
# 響應大小(kernel size)參數為 apertureSize 預設(3,3)
cannyKey = cv.Canny(gray_key, 50, 150)
cannyLena = cv.Canny(gray_lena, 50, 150)

#顯示影像處理結果並與原圖比較
cv.imshow('Ori_key', img_key) 
cv.imshow('Ori_Lena',img_lena)
cv.imshow('key_canny',cannyKey)
cv.imshow('lena_canny',cannyLena)

cv.waitKey()
cv.destroyAllWindows()
```
