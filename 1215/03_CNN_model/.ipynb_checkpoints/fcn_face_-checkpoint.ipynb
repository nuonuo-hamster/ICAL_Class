```python
from rcnn.load_data import rcnn_format
import numpy as np
# 載入資料集
(train_x, train_y), (test_x, test_y) = rcnn_format('data2', (92, 112, 3))
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)

print(train_y)
print(test_y)
```

```python
# 檢查輸入是否已經normalize
print(np.max(train_x))
print(np.min(train_x))
print(np.max(test_x))
print(np.min(test_x))
```

```python
from keras.callbacks import TensorBoard
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Flatten,Dense
from keras.utils import np_utils
```

```python
# detection format to one-hot encoding function
def to_categorical(y):
    label_y = y[:,0]
    categorical_y = np_utils.to_categorical(label_y)
    print(label_y)
    print(categorical_y)
    return categorical_y
```

```python
# One-hot encoding
train_y = to_categorical(train_y)
test_y = to_categorical(test_y)
```

```python
model = Sequential()
model.add(Conv2D(filters=8, kernel_size=(5, 5), padding='same', activation='relu', input_shape=(112, 92, 3)))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(filters=8, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(filters=8, kernel_size=(3, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(5 , activation='softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy', \
              optimizer='adam',                 \
              metrics=['accuracy'])
```

```python
# 訓練模型
history = model.fit(x=train_x, y=train_y, epochs = 20, batch_size=20)
```

```python
# 畫出loss圖形
import matplotlib.pyplot as plt

loss = history.history['loss']
epochs = len(loss)
plt.plot(range(epochs), loss, marker='.', label='loss(training data)')
plt.legend(loc='best')
plt.grid()
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
model.save('fcn.h5')
```

```python
from tensorflow.keras.models import load_model
model = load_model('fcn.h5')
```

```python
# 測試
model.evaluate(x=test_x, y=test_y)
```

```python
# 預測
predict = model.predict(x=test_x)
print(predict)
print(test_y)
```

```python
# 顯示圖片及預測結果的函數
import random

def plot_images_labels_prediction(images, labels, prediction=[], names=[], num=25):
    # 取得 pyplot 物件參考
    fig = plt.gcf()

    # 設定畫布大小為 12 吋*14吋
    fig.set_size_inches(12, 14)

    idx = random.randint(0, images.shape[0] - 1)

    # 限制最多顯示 25 個子圖
    if num > 25: num = 25 

    # 依序顯示 num 個子圖
    for i in range(0, num):
        # 建立 5*5 個子圖中的第 i+1 個 
        ax = plt.subplot(5, 5, i + 1)

        # 顯示子圖
        ax.imshow(images[idx], cmap='binary')

        title = ""
        color = 'black'
        label = labels[idx]
        pred = None

        # 有預測值就加入標題中
        if len(prediction) > 0:
            if prediction[idx] != labels[idx]:
                color = 'red'
            pred = prediction[idx]
        
        if len(names) > 0:
            label = names[label]
            if pred != None:
                pred = names[pred]

        title = "label = " + str(label)
        if pred != None: title += "\npredict = " + str(pred)


        #設定標題
        ax.set_title(title, fontsize=10, color=color) 
        
        #不顯示 x 、y軸刻度
        ax.set_xticks([]);  
        ax.set_yticks([]);
        
        #樣本序號增量 1
        idx += 1    
        if idx >= images.shape[0]:
            idx = 0
    
    #繪製圖形
    plt.subplots_adjust(wspace=0.6, hspace=0.6)
    plt.show()  

def plot_acc_val(acc, val):
    plt.plot(acc)
    plt.plot(val)
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```python
NAMES = ['A', 'B', 'C', 'D']
plot_images_labels_prediction(test_x, np.argmax(test_y, axis=1), np.argmax(predict, axis=1), NAMES)
```
