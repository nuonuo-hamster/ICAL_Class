# 灰階化

### 1. 載入模組

```python
import numpy as np
import cv2 as cv
```

### 2. 影像灰階化

```python
# 讀取圖檔
img = cv.imread("./img/polystyrene.jpg")
# img = cv.imread("./img/board.jpg")

# BGR轉灰階(OpenCV中預設影像讀取格式為BGR)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# 顯示影像資訊（寬、高、<灰階圖片通道數為1，所以只剩兩個數值>)
print("Shape of Color Image: {}".format(img_gray.shape))

# 顯示灰階影像
cv.imshow("result", img_gray)


#按下任意鍵關閉視窗
cv.waitKey(0)
cv.destroyAllWindows()
```
