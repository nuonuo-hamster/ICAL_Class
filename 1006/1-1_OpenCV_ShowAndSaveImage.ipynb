# OpenCV 影像讀取

### 1. 載入模組

```python
import numpy as np
import cv2 as cv
```

### 2. 顯示影像

<img src="./img/polystyrene.jpg" width=20%> 256\*330  <img src="./img/board.jpg" width=20%> 256\*330

```python
# 讀取影像
img = cv.imread("./img/polystyrene.jpg")
# img = cv.imread('./img/board.jpg')

# 顯示影像資訊（寬、高、通道數 <RGB彩色圖片為3,灰階圖片為1>)
print ("Width: {}\nHeight: {}\nChannel: {}".format(img.shape[0],img.shape[1],img.shape[2]))

# 顯示圖片 (名稱, 圖片變數)
cv.imshow("Image", img)

# 按下任意鍵關閉視窗 (記得要加上destroyAllWindows()，否則程式無法正常結束)
cv.waitKey(0)
cv.destroyAllWindows()
```

### 3. 儲存影像

```python
# 儲存圖片 (名稱, 圖片變數)
cv.imwrite("./img/polystyrene_copy.jpg", img)
# cv.imwrite("./img/board_copy.jpg", img)
```
