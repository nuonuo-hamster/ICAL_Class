```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image


image = cv2.imread("/content/drive/MyDrive/hw/Lenna.jpg")
# image = cv2.imread("/content/drive/MyDrive/hw/ball.jpg")

#轉換成lab空間
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)
#切割通道
l,a,b = cv2.split(lab_image)

#將BGR轉換成RGB空間
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
lab_image = cv2.cvtColor(lab_image, cv2.COLOR_BGR2RGB)

#顯示原始圖像和lab轉換後的圖像
plt.figure(1)
plt.subplot(121)
plt.imshow(image)
plt.title("img")
plt.subplot(122)
plt.imshow(lab_image)
plt.title("lab_image")
plt.show()

#顯示l,a,b通道
plt.figure(1)
plt.subplot(131)
plt.imshow(l)
plt.title("l")
plt.subplot(132)
plt.imshow(a)
plt.title("a")
plt.subplot(133)
plt.imshow(b)
plt.title("b")
plt.show()
```

```python
import cv2
from python_color_transfer.color_transfer import ColorTransfer

# Using demo images
input_image = '/content/drive/MyDrive/hw/1.jpg'
ref_image = '/content/drive/MyDrive/hw/2.jpg'
# input_image = '/content/drive/MyDrive/hw/tuberculosis-phone-0009.jpg'
# ref_image = '/content/drive/MyDrive/hw/tuberculosis-phone-0328.jpg'

# input image and reference image
img_arr_in = cv2.imread(input_image)
img_arr_ref = cv2.imread(ref_image)

# Initialize the class
PT = ColorTransfer()

# Lab mean transfer
img_arr_lt = PT.lab_transfer(img_arr_in=img_arr_in, img_arr_ref=img_arr_ref)

img_arr_in = cv2.cvtColor(img_arr_in, cv2.COLOR_BGR2RGB)
img_arr_ref = cv2.cvtColor(img_arr_ref, cv2.COLOR_BGR2RGB)
img_arr_lt = cv2.cvtColor(img_arr_lt, cv2.COLOR_BGR2RGB)

plt.figure(1)
plt.subplot(131)
plt.imshow(img_arr_in)
plt.title("img_arr_in")
plt.subplot(132)
plt.imshow(img_arr_ref)
plt.title("img_arr_ref")
plt.subplot(133)
plt.imshow(img_arr_lt)
plt.title("img_arr_lt")
plt.show()

```
