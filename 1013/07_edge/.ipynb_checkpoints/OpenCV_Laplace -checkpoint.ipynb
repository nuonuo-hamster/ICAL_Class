# OpenCV_Laplacian

```python
import numpy as np
import cv2 as cv
```

```python
#載入Lena範例影像，並將其轉換為灰階影像
img_lena = cv.imread("./img/lena.png")
gray_lena = cv.cvtColor(img_lena, cv.COLOR_BGR2GRAY)

#載入鑰匙範例影像，並將其轉換為灰階影像
img_key = cv.imread("./img/key.png")
gray_key = cv.cvtColor(img_key, cv.COLOR_BGR2GRAY)

cv.imshow('Ori_key', img_key) 
cv.imshow('Ori_Lena',img_lena)  
cv.imshow('lena_gray',gray_lena)
cv.imshow('key_gray',gray_key)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Laplace 響應大小(kernel size)為 3*3
# ddepth: -1 表示影像深度與原圖相同
laplaceKey = cv.Laplacian(src=gray_key, ddepth=-1, ksize=3)
laplaceLena = cv.Laplacian(src=gray_lena, ddepth=-1, ksize=3)

cv.imshow('Ori_key', img_key) 
cv.imshow('Ori_Lena',img_lena)  
cv.imshow('key_laplace',laplaceKey)
cv.imshow('lena_laplace',laplaceLena)

cv.waitKey()
cv.destroyAllWindows()
```
