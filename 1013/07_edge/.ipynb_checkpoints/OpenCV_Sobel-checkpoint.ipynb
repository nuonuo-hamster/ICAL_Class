# OpenCV_Sobel

```python
import numpy as np
import cv2 as cv
```

```python
#載入Lena範例影像，並將其轉換為灰階影像
img_lena = cv.imread("lena.png")
gray_lena = cv.cvtColor(img_lena, cv.COLOR_BGR2GRAY)

#載入鑰匙範例影像，並將其轉換為灰階影像
img_key = cv.imread("key.png")
gray_key = cv.cvtColor(img_key, cv.COLOR_BGR2GRAY)


cv.imshow('Ori_Lena',img_lena)  
cv.imshow('lena_gray',gray_lena)
cv.imshow('Ori_key', img_key)  
cv.imshow('key_gray',gray_key)

cv.imwrite('gray_lena.png',gray_lena)
cv.imwrite('gray_key.png',gray_key)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Sobel 分別針對 x 與 y 方向響應
dx_key = cv.Sobel(src=gray_key, ddepth=cv.CV_16S, dx=1, dy=0, ksize=3)
dx_lena = cv.Sobel(src=gray_lena, ddepth=cv.CV_16S, dx=1, dy=0, ksize=3)

# Sobel 計算 x 與 y 方向響應介於-1020~1020(若使用 8-bits unsigned integer 無法表示)
dy_key = cv.Sobel(src=gray_key, ddepth=cv.CV_16S, dx=0, dy=1, ksize=3)
dy_lena = cv.Sobel(src=gray_lena, ddepth=cv.CV_16S, dx=0, dy=1, ksize=3)

# 因此此處使用 16-bits signed integer
# Sobel 將響應計算結果(-1020~1020)取絕對值後轉成 uint-8
Xaxis_key = cv.convertScaleAbs(src=dx_key)
Xaxis_lena = cv.convertScaleAbs(src=dx_lena)
Yaxis_key = cv.convertScaleAbs(src=dy_key)
Yaxis_lena = cv.convertScaleAbs(src=dy_lena)

cv.imshow('x_axis_key',Xaxis_key)
cv.imshow('x_axis_lena',Xaxis_lena)
cv.imshow('y_axis_key',Yaxis_key)
cv.imshow('y_axis_lena',Yaxis_lena)

cv.waitKey()
cv.destroyAllWindows()
```

```python
# Sobel 將 x 與 y 方向響應合併
sobel_key = cv.addWeighted(src1=Xaxis_key, alpha=0.5, src2=Yaxis_key,beta=0.5, gamma=0)
sobel_lena = cv.addWeighted(src1=Xaxis_lena, alpha=0.5, src2=Yaxis_lena,beta=0.5, gamma=0)
cv.imshow('Sobel_key',sobel_key)
cv.imshow('Sobel_Lena',sobel_lena)

cv.imwrite('SobelKey.png',sobel_key)
cv.imwrite('SobelLena.png',sobel_lena)

cv.waitKey()
cv.destroyAllWindows()
```
